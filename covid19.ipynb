# Covid Vaccination Rates Comparison Project

```python
# Relevant Imports

import numpy as np
import pandas as pd
import plotly.figure_factory as ff
import plotly.express as px
```

Import Data from online repository

```python
url = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'
world_data = pd.read_csv(url, header=0)
world_data.index = pd.to_datetime(world_data['date'])
world_data = world_data.sort_index()
countries = np.unique(world_data['location'])
```

```python
OECD = ["Australia", "Austria", "Belgium", "Canada", "Chile", "Colombia", "Costa Rica", "Czechia", "Denmark", "Estonia", 
        "Finland", "France", "Germany", "Greece", "Hungary", "Iceland", "Ireland", "Israel", "Italy", "Japan",
        "South Korea", "Latvia", "Lithuania", "Luxembourg", "Mexico", "Netherlands", "New Zealand", "Norway", "Poland", "Portugal",
        "Slovakia", "Slovenia", "Spain", "Sweden", "Switzerland", "Turkey", "United Kingdom", "United States"]
oecd_data = world_data[world_data['location'].isin(OECD)]
    
```

# Covid 19 Tests
Comparing Total Vaccinations per Thousand By Country


```python
def latest_vacc(data, metric):
    latest_vacc_rates = data.copy()
    latest_vacc_rates.index = latest_vacc_rates['location']
    latest_vacc_rates = latest_vacc_rates[pd.notnull(latest_vacc_rates[metric])]
    latest_vacc_rates = latest_vacc_rates.drop_duplicates("location", keep="last")
    return latest_vacc_rates.sort_index()
```

```python
oecd_vacc = latest_vacc(oecd_data, 'people_vaccinated_per_hundred')
vaccination_percentage = px.bar(oecd_vacc, x='location', y='people_vaccinated_per_hundred')
vaccination_percentage.write_image("out/vacc_percent.png")
vaccination_percentage
```

```python
oecd_total = oecd_data.query("date > '2021-08-01'")
total_fig = px.line(oecd_total, x="date", y="new_cases_smoothed_per_million", color='location', height=600, width=1000)
total_fig.write_image("out/cases_v_time.png")
total_fig
```

```python
vaccinated_data = oecd_data.query("date > '2021-01-01'")
vacc_v_time_plot = px.line(vaccinated_data, x="date", y='people_vaccinated_per_hundred', color='location', height=600, width=1000)
vacc_v_time_plot.write_image("out/vacc_v_time.png")
vacc_v_time_plot
```
