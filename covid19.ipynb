# Covid 19 Tests
Comparing Total Vaccinations per Thousand By Country


```python
def latest_vacc(data, metric):
    latest_vacc_rates = data.copy()
    latest_vacc_rates.index = latest_vacc_rates['location']
    latest_vacc_rates = latest_vacc_rates[pd.notnull(latest_vacc_rates[metric])]
    latest_vacc_rates = latest_vacc_rates.drop_duplicates("location", keep="last")
    return latest_vacc_rates.sort_index()
```

```python
oecd_vacc = latest_vacc(oecd_data, 'people_vaccinated_per_hundred')
vaccination_percentage = px.bar(oecd_vacc, x='location', y='people_vaccinated_per_hundred')
vaccination_percentage.write_image("Graphs/people_vaccinated_per_hundred.png")
vaccination_percentage
```

```python
def plot_countries(data, metric, query=None, name="output"):
    if query is not None:
        df = data.query(query)
    else:
        df = data
    fig = px.line(df, x="date", y=metric, color='location', height=600, width=1000)
    fig.write_image(f"Graphs/{metric}.png")
    return fig
```
