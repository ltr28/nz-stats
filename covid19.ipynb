```python
world_data = pd.read_csv("owid-covid-data.csv", header=0)
world_data.index = pd.to_datetime(world_data['date'])
world_data = world_data.sort_index()
countries = np.unique(world_data['location'])
world_data[world_data['location']=="New Zealand"]['new_tests_smoothed_per_thousand'].tail(21)
```

```python
OECD = ["Australia", "Austria", "Belgium", "Canada", "Chile", "Colombia", "Costa Rica", "Czechia", "Denmark", "Estonia", 
        "Finland", "France", "Germany", "Greece", "Hungary", "Iceland", "Ireland", "Israel", "Italy", "Japan",
        "South Korea", "Latvia", "Lithuania", "Luxembourg", "Mexico", "Netherlands", "New Zealand", "Norway", "Poland", "Portugal",
        "Slovakia", "Slovenia", "Spain", "Sweden", "Switzerland", "Turkey", "United Kingdom", "United States"]
def oecd_countries(world):
    oecd_data = world[world['location'].isin(OECD)]
    return oecd_data
oecd_data = oecd_countries(world_data)
    
```

# Covid 19 Tests
Comparing Total Vaccinations per Thousand By Country


```python
def latest(data, metric):
    latest_vacc_rates = data.copy()
    latest_vacc_rates.index = latest_vacc_rates['location']
    latest_vacc_rates = latest_vacc_rates[pd.notnull(latest_vacc_rates[metric])]
    latest_vacc_rates = latest_vacc_rates.drop_duplicates("location", keep="last")
    return latest_vacc_rates.sort_index()
```

```python
oecd_vacc = latest(oecd_data, 'people_vaccinated_per_hundred')

px.bar(oecd_vacc, x='location', y='people_vaccinated_per_hundred')
```

```python
def plot_countries(data, metric, query=None):
    if query is not None:
        df = data.query(query)
    else:
        df = data
    fig = px.line(df, x="date", y=metric, color='location', height=800, width=1500)
    return fig
```

```python
plot_countries(oecd_data, "new_cases_smoothed_per_million")
```

```python

vaccinated_data = oecd_data.query("date>'2021'")
plot_countries(vaccinated_data, 'people_vaccinated_per_hundred')
```

```python
plot_countries(oecd_data, 'new_tests_smoothed_per_thousand')
```
